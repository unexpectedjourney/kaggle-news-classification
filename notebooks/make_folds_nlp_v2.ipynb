```python
from pathlib import Path

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import KBinsDiscretizer

%config InlineBackend.figure_format = 'retina' 
```

```python
random_state = 42
max_length = 512
```

```python
data_dir = Path("data")
folds_dir = Path("folds") / "nlp-v2"
train_path = data_dir / "train.csv"
```

```python
df = pd.read_csv(train_path, index_col="Id")
```

```python
df.head()
```

## Initial distribution

```python
sns.displot(df["source"])
```

## Stratify

```python
kfold = StratifiedKFold(n_splits=5, random_state=random_state, shuffle=True)  
```

```python
target_values = df[["source"]]
# est = KBinsDiscretizer(n_bins=50, encode='nominal', strategy='quantile')
# stratify_on = est.fit_transform(target_values).T[0]
splits = kfold.split(df, target_values)
```

```python
splits
```

```python
folds = []
for _, test_indexes in splits:
    folds.append(df.iloc[test_indexes])
```

```python
for fold in folds:
    print(max(fold["source"]))
    sns.displot(fold["source"])
    plt.show()
```

## Split sentence and extend group

```python
def extend_folds_by_text(fold):
    titles = []
    texts = []
    sources = []

    for i in range(fold.shape[0]):
        item = fold.iloc[i]
        source = item.source
        title = item.title
        text_lst = item.text.split()
        for num in range(len(text_lst) // max_length):
            titles.append(title)
            sources.append(source)
            texts.append(" ".join(text_lst[num * max_length:(num + 1) * max_length]))
        if len(text_lst) % max_length > 0:
            titles.append(title)
            sources.append(source)
            texts.append(" ".join(text_lst[(len(text_lst) // max_length) * max_length:(len(text_lst) // max_length + 1) * max_length]))

    new_fold = pd.DataFrame({
        "title": titles,
        "text": texts,
        "source": sources
    })
    
    return new_fold
```

```python
new_folds = []
for fold in folds:
    new_folds.append(extend_folds_by_text(fold))
```

```python
folds = new_folds
```

## Post balancing

```python
def upsample(fold, max_value):
    print(fold.shape)
    for label in fold.source.unique():
        values_to_add = max_value - fold[fold.source == label].shape[0]
        
        additional_fold_data = pd.DataFrame()
        
        while values_to_add > 0:
            for i in range(fold.shape[0]):
                if values_to_add == 0:
                    break

                if fold.iloc[i]["source"] == label:
                    additional_fold_data = additional_fold_data.append(fold.iloc[i])
                    values_to_add -= 1
        fold = fold.append(additional_fold_data)
    print(fold.shape)
    return fold
```

```python
np.max(folds[0].source.value_counts().values)
```

```python
b_folds = []
for fold in folds:
    b_folds.append(upsample(fold, np.max(fold.source.value_counts().values)))
```

```python
folds = b_folds
```

```python
for fold in folds:
    print(max(fold["source"]))
    sns.displot(fold["source"])
    plt.show()
```

```python
for i, fold in enumerate(folds):
    fold.to_csv(folds_dir / f"fold_{i}.csv", index=False)
```
